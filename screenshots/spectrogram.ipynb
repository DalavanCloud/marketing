# Simple spectral analysisAn illustration of the [Discrete Fourier Transform](http://en.wikipedia.org/wiki/Discrete_fourier_transform)$$ X_k = \sum_{n=0}^{N-1} x_n e^{-\frac{2 \pi i}{N} k n} \quad \quad k = 0, \dots, N-1 $$</br>using windowing, to reveal the frequency content of a sound signal.We begin by loading a datafile using SciPy's audio file support:

```python
from scipy.io import wavfilerate, x = wavfile.read('test_mono.wav')
```

And we can easily view its spectral structure using matplotlib's builtin ``specgram`` routine:

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))ax1.plot(x); ax1.set_title('Raw audio signal')ax2.specgram(x); ax2.set_title('Spectrogram');
```
